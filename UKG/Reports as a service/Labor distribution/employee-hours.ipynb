```python
#Rename the ID column
time_df.rename(columns={'Id': 'TimeId'}, inplace=True)
```

```python
list(time_df['EmpId'].unique())
```

```python
#Create a short and long emp id column. We will use the short id column for the merge and then drop it later

time_df['ShortEmpId'] = time_df['EmpId'].apply(lambda x: str(x)[5:9])
time_df['LongEmpId'] = time_df['EmpId'].apply(lambda x: str(x)[0:9])

time_df
```

## Bring in the timesheet endpoint

```python
#Bring in the timesheet endpoint to get the pay period
url="https://kN21.ultipro.com/UtmOdataServices/api/Timesheet?$filter=year(Starts) eq " + year + " and month(Starts) eq " + month
#Our username and password for the API authentication
username = "znarkin"
password = "page!i)yCI5-Y26o}7i("


#The response variable
response2 = requests.get(url, auth=HTTPBasicAuth(username, password))
```

```python
data2 = response2.json()
#data2
```

```python
timesheet_df = pd.DataFrame(data2['value'])
timesheet_df
```

```python
timesheet_df.columns
```

```python
#Subset the columns on timesheet_df
timesheet_df = timesheet_df[['EmployeeId', 'Starts', 'Ends', 'Id']]
timesheet_df
```

```python
#Rename the starts and ends columns
timesheet_df.rename(columns={'Starts': 'PayPeriodStart', 'Ends': 'PayPeriodEnd', 'EmployeeId': 'EmpId', 'Id': 'TimesheetId'}, inplace=True)
timesheet_df
```

```python
#Update the EmpId column so that is is string datatype
timesheet_df['EmpId'] = timesheet_df['EmpId'].astype('str')
time_df['EmpId'] = time_df['EmpId'].astype('str')
```

```python
#Update the dates for pay period start and end
timesheet_df['PayPeriodStart'] = timesheet_df['PayPeriodStart'].apply(lambda x: str(x)[:10])
timesheet_df['PayPeriodEnd'] = timesheet_df['PayPeriodEnd'].apply(lambda x: str(x)[:10])

timesheet_df
```

## Join the time and timesheet dataframes

```python
merged_df = time_df.merge(timesheet_df, left_on=['ShortEmpId', 'WorkDate'], right_on=['EmpId')
merged_df
```
